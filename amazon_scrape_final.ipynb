```python
brand  = []
title  = []
price  = []
rating = []
review = []
for URL_len in range(len(URL)): 
    webpage = None

    while not webpage or webpage.status_code != 200:
     
        webpage = requests.get(URL[URL_len], headers=HEADERS)

    print(webpage)


        
    soup = BeautifulSoup(webpage.content, "html.parser")

        
    divs = soup.find_all('div', class_ = "sg-col-4-of-24 sg-col-4-of-12 s-result-item s-asin sg-col-4-of-16 sg-col s-widget-spacing-small sg-col-4-of-20")



    for i in range(len(divs)):
        
        element = divs[i].find('div', class_="a-section a-spacing-small puis-padding-left-micro puis-padding-right-micro").find('div', class_="a-section a-spacing-none a-spacing-top-small s-title-instructions-style").find('div', class_="a-row a-size-base a-color-secondary").find('h2', class_="a-size-mini s-line-clamp-1").find('span', class_="a-size-base-plus a-color-base")
        if element:
            brand.append(element.text.strip())
        #title
        element_1 = divs[i].find('div', class_="a-section a-spacing-small puis-padding-left-micro puis-padding-right-micro").find('div', class_="a-section a-spacing-none a-spacing-top-small s-title-instructions-style").find('h2', class_="a-size-mini a-spacing-none a-color-base s-line-clamp-2").find('a', class_="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal").find('span', class_="a-size-base-plus a-color-base a-text-normal")
        if element_1:
            title.append(element_1.text.strip())
        else:
            title.append("none")

        element_2 = divs[i].find('div', class_="a-section a-spacing-small puis-padding-left-micro puis-padding-right-micro").find('div', class_="a-section a-spacing-none a-spacing-top-small s-price-instructions-style").find('span', class_="a-price-whole")
        if element_2:
            price.append(element_2.text.strip())
        else:
            price.append("none")
        # rating
        rating_classes = ["a-star-small-4", "a-star-small-3-5", "a-star-small-4-5","a-star-small-5","a-star-small-3","a-star-small-2-3","a-star-small-2","a-star-small-1-2","a-star-small-1"]
        rating_1 = []

        for class_name in rating_classes:
            element_3 = divs[i].find('div', class_="a-section a-spacing-small puis-padding-left-micro puis-padding-right-micro").find('i', class_=f"a-icon a-icon-star-small {class_name} aok-align-bottom")

            if element_3:
                rating_1.append(element_3.text.strip())
                break

        if not rating_1:
            rating.append('none')
        else:
            rating.append(rating_1)
        
        try:
            element_4 = divs[i].find('div', class_="a-section a-spacing-small puis-padding-left-micro puis-padding-right-micro").find('div', class_="a-section a-spacing-none a-spacing-top-micro").find('span', class_="a-size-base s-underline-text")
            if element_4:
                review.append(element_4.text.strip())
            else:
                review.append("none")
        except AttributeError:
            review.append("none")
```

```python
amazon_df = pd.DataFrame()
amazon_df['title']  = title
amazon_df['brand']  = brand
amazon_df['price']  = price
amazon_df['rating'] = rating
amazon_df['review'] = review
```
